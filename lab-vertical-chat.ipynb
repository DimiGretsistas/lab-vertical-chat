```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
import panel as pn

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

```python
#Creating the prompt
#read and understand it.
import panel as pn  # GUI

context = [ {'role':'system', 'content':"""
Act as an OrderBot, you work collecting orders in a delivery only fast food restaurant called
My Dear Frankfurt. \
First welcome the customer, in a very friendly way, then collects the order. \
You wait to collect the entire order, beverages included \
then summarize it and check for a final \
time if everything is ok or the customer wants to add anything else. \
Finally you collect the payment.\
Make sure to clarify all options, extras and sizes to uniquely \
identify the item from the menu.\
You respond in a short, very friendly style. \
The menu includes \
burger  12.95, 10.00, 7.00 \
frankfurt   10.95, 9.25, 6.50 \
sandwich   11.95, 9.75, 6.75 \
fries 4.50, 3.50 \
salad 7.25 \
Toppings: \
extra cheese 2.00, \
mushrooms 1.50 \
martra sausage 3.00 \
canadian bacon 3.50 \
romesco sauce 1.50 \
peppers 1.00 \
Drinks: \
coke 3.00, 2.00, 1.00 \
sprite 3.00, 2.00, 1.00 \
vichy catalan 5.00 \
"""} ]

#Creating the panel.
pn.extension()

panels = []

client_prompt = pn.widgets.TextInput(value="Hi", placeholder='Enter text here…')
button_conversation = pn.widgets.Button(name="talk")

interactive_conversation = pn.bind(add_prompts_conversation, button_conversation)

dashboard = pn.Column(
    client_prompt,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True),
)

dashboard
```

*Report*

In this exercise, I tested three different chatbot prompt designs using OpenAI, Python, and Panel to understand how prompt structure affects response quality.

The first model was a cleaning service chatbot. This version worked well because the business type was clear, the available services were specific, and the chatbot had one focused task: collect customer inquiries. It consistently asked for relevant details such as service type, address, date, and contact information. Responses were structured and reliable.

The second model was a personal training studio chatbot. This version also performed well. Even though it offered more services and options, all services still belonged to one industry. Because the prompt clearly defined goals, packages, fitness levels, and booking steps, the chatbot remained accurate and helpful. It handled bookings in a natural and professional way. There was some inconvnience when it offeres online-coaching evern though i booked in-person personal training.

The third model was designed very broadly. It combined many unrelated services such as fitness coaching, food delivery, car repair, cleaning, tutoring, and travel planning in one company. This caused confusion because the chatbot had no clear business identity or workflow. Some prices were vague, service options were unclear, and instructions conflicted with each other. In some cases, GPT gave incomplete answers, guessed missing information (as advides to try the same), or struggled to decide which questions to ask first.

Overall, the test showed that focused prompts with clear roles, limited scope, and structured instructions lead to much better chatbot performance than broad or conflicting prompts.

*What did I learn?*

Clear prompts create more accurate and consistent responses.
One business type works better than many unrelated services.
Defined workflows help the chatbot ask the right questions.
Too many options increase confusion and weak answers.
Ambiguous pricing or unclear categories can cause hallucinations.
Prompt engineering is essential for useful chatbot behavior.
